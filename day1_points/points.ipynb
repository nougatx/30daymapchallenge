```python
import folium
import geopandas as gpd
from branca.element import Template, MacroElement

# Load your GeoJSON file
geojson_path = 'query1.geojson'  # Replace with your actual GeoJSON file path
mosques = gpd.read_file(geojson_path)

# Create a folium map centered around Abu Dhabi
m = folium.Map(location=[24.4539, 54.3773], zoom_start=12)

# Function to color points based on the presence of a female section
def color_female_section(femaleroom):
    if femaleroom == 1:
        return 'purple'  # Mosques with female section
    else:
        return '#0073d3'  # Mosques without female section

# Add mosque points to the map with different colors
for _, mosque in mosques.iterrows():
    folium.CircleMarker(
        location=[mosque.geometry.y, mosque.geometry.x],  # GeoJSON coordinates
        radius=5,
        color=color_female_section(mosque['FEMALEROOM']),
        weight =1,
        opacity = 0.6,
        fill=True,
        fill_color=color_female_section(mosque['FEMALEROOM']),
        fill_opacity=0.6,
        popup=folium.Popup(f"Mosque: {mosque['NAMEENG']}, Female Section: {'Yes' if mosque['FEMALEROOM'] == 1 else 'No'}")
    ).add_to(m)

# Create a custom legend using HTML/CSS
legend_html = '''
<div style="
    position: fixed; 
    top: 50px; left: 50px; width: 155px; height: 120px; 
    z-index:9999; font-size:14px;
    background-color:white; opacity: 0.7;
    ">
    <p style="margin: 8px;"><strong>Mosque Locations, Abu Dhabi Island</strong></p>
    <i style="background: purple; width: 10px; height: 10px;border-radius: 50%; display: inline-block; margin-right: 5px;"></i>With Female Section<br>
    <i style="background: #0073d3; width: 10px; height: 10px;border-radius: 50%; display: inline-block; margin-right: 5px;"></i>Without Female Section<br>
</div>
'''

# Add the legend to the map
m.get_root().html.add_child(folium.Element(legend_html))

# Save map to HTML or display
m.save('mosques_in_ad.html')
display(m)
```
